### Carregando e visualizando os dados

### Tratamento de dados

```python
# verificando como está o dado referente ao valor da mensalidade

df_dados['VLMENSAL'].value_counts()
```

```python
# Muda vírgula por ponto para transformar em float

df_dados['VLMENSAL'] = df_dados['VLMENSAL'].str.replace(',', '.')
df_dados['VLMENSAL'].value_counts()
```

```python
df_dados['VLMENSAL'] = df_dados['VLMENSAL'].astype(float)
print('Tipo de dado após alteração: ',df_dados['VLMENSAL'].dtypes)
df_dados['VLMENSAL'].value_counts()
```

```python
# Filtra somente casos com valor mensal (recorrência)

df_recorrentes = df_dados[df_dados['VLMENSAL'] > 0]
df_recorrentes.head(5)
```

```python
df_recorrentes.info()
```

```python
df_recorrentes.isnull().sum()
```

### Explorando os dados

```python
# Separa DataFrame em ativos e cancelamentos

df_cancelados = df_recorrentes[df_recorrentes['CANCELAMENTO'].notnull()]
df_ativos = df_recorrentes[df_recorrentes['CANCELAMENTO'].isnull()]

print('Qtd registros de ativos: ', df_ativos.size, ' -> ', round((df_ativos.size/df_recorrentes.size)*100,2),'%')
print('Qtd registros de cancelados: ', df_cancelados.size, ' -> ', round((df_cancelados.size/df_recorrentes.size)*100,2),'%')
```
