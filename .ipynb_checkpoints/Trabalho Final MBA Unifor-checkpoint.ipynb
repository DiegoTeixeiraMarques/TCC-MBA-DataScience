```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(open(r'dataset.xlsx', 'rb'), sheet_name='Planilha1', engine='openpyxl')
df
```

```python
df.info()
```

```python
df1 = df[df.VLMENSAL > 0]
df1.head(5)
```

```python
churn = []
serieCancelados = df1['CANCELAMENTO']
for i in serieCancelados:
    #print(i)
    if i is not pd.NaT:
        churn.append(1)
    else:
        churn.append(0)
print('Qtd Churn: ', len(churn), '\n')
print('Qtd serieCancelados: ', len(serieCancelados))
```

```python
target = pd.DataFrame(churn, columns=['churn'])
target
```

```python
df1.columns
```

```python
features = ['NMCLIENTE', 'UF', 'NMMODALIDADE', 'NMSEGMENTO', 'NMMERCADO', 
       'NMMOTIVOCANCELAMENTO', 'TEMPODIAS', 'VLMENSAL', ]
```

```python
df2 = df1[features]
df2.head(5)
```

### Analisando Features

```python
print('** Linhas, Colunas ** \n', df2.shape, '\n','-'*32,'\n')
print('** Tipos de dados ** \n', df2.dtypes, '\n','-'*32,'\n')
print('** Qtd de Nulos ** \n', df2.isnull().sum(), '\n','-'*32,'\n')

# Verificando quantidade de categorias por feature
print('** Categorias por Feature **')
for i in df2.columns:
    print(i, ' :', len(df2[str(i)].unique()))
```

### Analisando Target

```python
cancelado, naoCancelado = int(target[target.churn == 0].count()), int(target[target.churn == 1].count())
total = target.size

proporcaoTarget = pd.DataFrame({
    'Qtd de linhas': {'Cancelado': cancelado, 'Não Cancelado': naoCancelado},
    '%': {'Cancelado': round((cancelado / total)*100,2), 'Não Cancelado': round((naoCancelado / total)*100,2)}
})

proporcaoTarget
```

```python
modalidades = list(set(df2.NMMODALIDADE))
segmentos = list(set(df2.NMSEGMENTO))  
print(modalidades, '\n\n', segmentos)
```

```python
registrosModalidades = list(df2.NMMODALIDADE)
registrosSegmentos = list(df2.NMSEGMENTO)
contagemPorModalidades = []
contagemSegmentos = []

for i in modalidades:
    contagemPorModalidades.append(registrosModalidades.count(i))
for i in segmentos:
    contagemSegmentos.append(registrosSegmentos.count(i))
    
print(contagemPorModalidades, '\n\n', contagemSegmentos)    
```

```python
df_Modalidades = pd.DataFrame(data=modalidades, columns=['Modalidade'])
df_Modalidades['qtd'] = contagemPorModalidades

df_Segmentos = pd.DataFrame(data=segmentos, columns=['Segmentos'])
df_Segmentos['qtd'] = contagemSegmentos

df_Modalidades = df_Modalidades.sort_values(by=['qtd'], ascending=True)
df_Segmentos = df_Segmentos.sort_values(by=['qtd'], ascending=True)
```

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.barh(df_Modalidades.Modalidade, df_Modalidades.qtd, color='green')
plt.title('Quantidade de registros por modalidade')
```

```python
plt.rcParams['figure.figsize'] = (10,20)
plt.barh(df_Segmentos.Segmentos, df_Segmentos.qtd, color='green')
plt.title('Quantidade de registros por segmentos')
```
